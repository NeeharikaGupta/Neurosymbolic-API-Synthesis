```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Stanford NLP/dataset.csv")

df_few_shot= []

for idx, row in df.iterrows():
    prompt = (
        f"{row['Few shot t1'].strip()}\n"
        f"{row['Few shot c1'].strip()}\n\n"
        f"{row['Few shot t2'].strip()}\n"
        f"{row['Few shot c2'].strip()}\n\n"
        f"{row['Task'].strip()}\n"
    )
    df_few_shot.append(prompt)
```

```python
output_path = "/content/drive/MyDrive/Stanford NLP/few_shot_prompts.csv"
new_df = pd.DataFrame({'prompt': df_few_shot})
new_df.to_csv(output_path, index=False)
```
