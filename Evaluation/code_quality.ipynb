````python
import pandas as pd
import tempfile
import subprocess
import re
````

````python
def extract_python_code(text):

    code_blocks = re.findall(r"```(?:python)?(.*?)```", text, re.DOTALL)
    code_blocks = [code.strip() for code in code_blocks]
    return code_blocks
````

````python
def run_pylint_on_code(code_snippet):
    with tempfile.NamedTemporaryFile(suffix='.py', mode='w', delete=False) as temp_file:
        temp_file.write(code_snippet)
        temp_file_name = temp_file.name
    try:
        result = subprocess.run(
            ['pylint', temp_file_name],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        return result.stdout
    except Exception as e:
        return f"Error running Pylint: {e}"
````

##Gemini

````python
for index, row in df.iterrows():
    print(f"Pylint output for row {index}")
    output = run_pylint_on_code(extract_python_code(row['direct_response_gemini'])[0])
    print(output)
````
